```python
print('Create a table with the 5-number summary of all the numeric attributes after treating missing values. Do you think the strategy used in dealing with missing values was effective?')
import pandas as pd
df=pd.read_csv("Cereals.csv")
df.describe() 
```

```python
print('For each of the numeric attributes (proteins up to vitamins), identify and replace all noisy data with the median of attribute.')
medians = df.median()
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df['lower_bound'] = Q1 - 1.5 * IQR
df['upper_bound'] = Q3 + 1.5 * IQR

for column in df.columns[3:12]:
    is_outlier = (df[column] < df['lower_bound']) | (df[column] > df['upper_bound'])
    df2=df.replace(is_outlier,medians)

df2
```

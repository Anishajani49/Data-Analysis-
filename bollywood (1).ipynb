```python
import pandas as pd
print('How many movies in each genre got released in different release times like long weekend, festive season, etc.')
df = pd.read_csv("/Users/anishajani/Downloads/bollywood.csv")
pd.crosstab(df['Genre'], df['ReleaseTime'])
```

```python
df['month'] = pd.DatetimeIndex(df['Release Date']).month
print('Which month of the year, maximum number movie releases are seen?')
movie=df.groupby('month')['MovieName']
print(movie.count())
print("month number with maximum releases:1")

```

```python
print('Which month of the year typically sees most releases of high budgeted movies, that is, movies with budget of 25 crore or more?')
df.query("Budget>=25")["month"].max()
```

```python
print('Which are the top 10 movies with maximum return on investment (ROI)? Calculate return on investment (ROI) as (BoxOfficeCollection – Budget) / Budget.')
n = df['BoxOfficeCollection'].sub(df['Budget'])
df['ROI']=n/df['Budget']
df.sort_values(by=['ROI']).head(10)
```

```python
print("8. Do the movies have higher ROI if they get released on festive seasons or long weekend? Calculate the average ROI for different release times.")
print(df.loc[df["ReleaseTime"] == "FS", "ROI"].mean())
print(df.loc[df["ReleaseTime"] == "LW", "ROI"].mean())
print("From the above calculations, movies have higher ROI if they get released on long weekends")
```

```python
import seaborn as sns
print("9. Is there a correlation between box office collection and YouTube likes? Is the correlation positive or negative?")
print(df[["BoxOfficeCollection", "YoutubeLikes"]].corr())
sns.pairplot(df[["YoutubeLikes", "BoxOfficeCollection"]], kind="reg")
print("\nThere is a positive correlation between the box office collection and youtube likes.")
```

```python
print("10. Which genre of movies typically sees more YouTube likes? Draw boxplots for each genre of movies to compare.")
print("The genre with most number of YouTube likes is: " + df.groupby(["Genre"])["YoutubeLikes"].sum().idxmax())
sns.boxplot(x="Genre", y="YoutubeLikes", data=df)
```

```python
print("11. Which of the variables among Budget, BoxOfficeCollection, YoutubeView, YoutubeLikes, YoutubeDislikes are highly correlated? Note: Draw pair plot or heatmap.")
sns.pairplot(df[["Budget", "BoxOfficeCollection", "YoutubeViews", "YoutubeLikes", "YoutubeDislikes"]], kind='reg')
print("YoutubeLikes and YoutubeDislikes are highly correlated")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
print("12. During 2013−2015 period, highlight the genre of movies and their box office collection? Visualize with best fit graph.")
sns.barplot(x='Release Date', y='BoxOfficeCollection', hue='Genre', data=df)
plt.show()
```

```python
print("13. Visualize the Budget and Box office collection based on Genre.")
plt.bar(df["Genre"],df["BoxOfficeCollection"],align='edge',width=0.5,label="BoxofficeCollection")
plt.bar(df["Genre"],df["Budget"],label="Budget")
plt.legend
plt.show()
```

```python
print("14. Find the distribution of movie budget for every Genre.")
print(df.groupby(df["Genre"])['Budget'].sum())

df.groupby('Genre')["Budget"].plot(kind='density')
plt.legend()
plt.show()
```

```python
print(" During 2013−2015, find the number of movies released in every year. Also, visualize with best fit graph.")
from datetime import datetime
import datetime as dt
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df['Release Date'] = pd.to_datetime(df['Release Date'])
temp_df = pd.DataFrame(df['Release Date'].dt.year.value_counts())
temp_df = temp_df.reset_index()
temp_df.columns = ['Year', 'Movie Count']
sns.barplot(x='Year', y='Movie Count', data=temp_df)
plt.xlabel('Year')
plt.ylabel('Number of Movies Released')
plt.title('Number of Movies Released per Year')
plt.show()
```
